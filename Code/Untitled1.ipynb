# Getting tags from each article.

Find most common words in each article, exclude Englsh stopwords, numbers and months names.

```python
from nltk.corpus import stopwords 
stop_words = list(stopwords.words('english')) 
file = open("months_list.txt", "r")
months_list = file.read().splitlines() 
```

Create function for filtering text.

```python
import re

def filter_text(text, stop_words=stop_words, others=months_list):
    result_text = []
    for word in re.split("\W+",text):
        if len(word) > 1:
            word = word.lower()
            if word not in (stop_words+others) and not word.isdigit():
                result_text.append(word)
    return result_text
```

```python
from collections import Counter 
  
def add_tags(data_frame):
    tags = []
    for article in data_frame['text']:
        article = filter_text(article)
        counter = Counter(article) 
        # most_common() produces k frequently encountered 
        # input values and their respective counts. 
        most_occur = counter.most_common(5) 
        most_occur = [words[0] for words in most_occur]
        tags.append(most_occur)
    
    data_frame['tags'] = tags
    return data_frame

fake_news = add_tags(fake_news)
fake_news
```

```python
from Levenshtein import distance as leve_dist
from functools import reduce

def get_distance(tags1, tags2):
    distance = 0
    for tag in tags1:
        distance += min([leve_dist(curr_tag, tag) for curr_tag in tags2])
    return distance 

def get_centroid(cluster_tags):
    counter = Counter(cluster_tags) 
    most_occur = counter.most_common(5) 
    centroid = [tags[0] for tags in most_occur]
    return centroid
```
